```python
import matplotlib.pyplot as plt
import numpy as np
from IPython import display
display.set_matplotlib_formats("svg")
```

```python
#解决中文乱码
import matplotlib as mpl
mpl.rcParams['font.sans-serif'] = ['SimHei']
mpl.rcParams['font.serif'] = ['SimHei']
mpl.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题,
```

```python


fig = plt.figure(figsize=(5.6, 7.2), dpi=100)
ax = fig.subplots()
ax.set_ylim(0,14000)
ax1 = ax.twinx()
ax1.set_ylim(0,3000)
x = np.arange(1997,2021)
y1 = np.array([252, 262, 322, 502, 718, 882, 1156, 1303, 1588, 1793, 1886, 2008, 2108, 2186, 2267, 2378, 2488, 2578, 2649, 2700, 2750, 2883, 3120, 3284])
y2 = np.array([5,15,25,35,45,56,66,77,87,97,107,117,127,137,147,159,168,177,186,196,206,216,226,236])
y3 = np.array([2000,2010,2030,2050,2350,2700,3100,3500,3700,3870,4380,4490,4600,4700,4810,4914,5000,5115,5225,5335,5445,5555,5670,5800])
ax.plot(x,y3,label = "学校",lw = 7)
ax.spines['right'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['bottom'].set_color('none')
ax.spines['top'].set_color('none')
ax1.spines['right'].set_color('none')
ax1.spines['left'].set_color('none')
ax1.spines['bottom'].set_color('none')
ax1.spines['top'].set_color('none')
x_label = list(x)
for i in range(len(x_label)):
    x_label[i] = str(x_label[i])
    x_label[i] += " 年"
ax1.scatter(x,y1,c="aqua",label = "学生 (万)",s = (x-1990)*5)
ax1.scatter(x,y2,c="orange",label = "教师 (万)",s = (x-1990)*5)
plt.legend(loc = 'upper left')
ax.legend(loc = 'upper right')
ax.set_xticks(x)
ax.set_xticklabels(x_label,rotation = 90)
ax.set_yticklabels(["0 K","2 K","4 K","6 K","8 K","10 K","12 K",""])
ax.hlines(2000,1997,2021,linestyles="dashed",color = "gray")
ax.hlines(4000,1997,2021,linestyles="dashed",color = "gray")
ax.hlines(6000,1997,2021,linestyles="dashed",color = "gray")
ax.hlines(8000,1997,2021,linestyles="dashed",color = "gray")
ax.hlines(10000,1997,2021,linestyles="dashed",color = "gray")
ax.hlines(12000,1997,2021,linestyles="dashed",color = "gray")
```
