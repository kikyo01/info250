```python
fig = plt.figure(figsize=(5.6, 7.2), dpi=100)
ax = fig.subplots()
ax.set_ylim(0,16000)
ax1 = ax.twinx()
ax1.set_ylim(-250,3500)
x = np.arange(1997,2021)
y1 = np.array([252, 262, 322, 502, 718, 882, 1156, 1373, 1588, 1793, 1886, 2008, 2108, 2186, 2267, 2368, 2458, 2546, 2618, 2680, 2750, 2883, 3120, 3384])
y2 = np.array([5,15,25,35,45,56,66,77,87,97,107,117,127,137,147,159,168,177,186,196,206,216,226,236])
y3 = np.array([2000,2010,2030,2040,2250,2645,3045,3445,3645,3820,4230,4380,4470,4570,4670,4774,4870,4965,5055,5145,5235,5325,5420,5510])
y4 = np.array([2100,2110,2230,2150,2410,2795,3215,3625,3845,4030,4510,4650,4760,4870,4980,5080,5190,5295,5405,5515,5625,5735,5850,5960])
ax.fill_between(x, y3 ,y4 ,color = 'blue',alpha=0.6, edgecolor ='none',label = "Colleges")
ax.spines['right'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['bottom'].set_color('none')
ax.spines['top'].set_color('none')
ax1.spines['right'].set_color('none')
ax1.spines['left'].set_color('none')
ax1.spines['bottom'].set_color('none')
ax1.spines['top'].set_color('none')
x_label = list(x)
for i in range(len(x_label)):
    x_label[i] = str(x_label[i])
ax1.scatter(x,y1,c="aqua",label = "Students (10K)",s = (x-1990)*15, alpha = 0.7)
ax1.scatter(x,y2,c="orange",label = "Teachers (10K)",s = (x-1990)*2.5, alpha = 0.7)
plt.legend(loc = 'upper left', frameon = 0)
ax.legend(loc = 'upper center', frameon = 0)
ax.set_xticks(x)
ax.set_xticklabels(x_label,rotation = 90)
ax.set_yticklabels(["0 K","2 K","4 K","6 K","8 K","10 K","12 K","14 K", ""])
ax.hlines(2000,1997,2021,linestyles="dashed",color = "gray")
ax.hlines(4000,1997,2021,linestyles="dashed",color = "gray")
ax.hlines(6000,1997,2021,linestyles="dashed",color = "gray")
ax.hlines(8000,1997,2021,linestyles="dashed",color = "gray")
ax.hlines(10000,1997,2021,linestyles="dashed",color = "gray")
ax.hlines(12000,1997,2021,linestyles="dashed",color = "gray")
ax.hlines(14000,1997,2021,linestyles="dashed",color = "gray")
ax.set_title('The development of Chinese education\n(1997-2020)')

fig.savefig(".\images\org.svg")
```
